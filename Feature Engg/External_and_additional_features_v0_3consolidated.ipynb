````python
def plot_bar(data, x_axis, y_axis, xlabel, ylabel):
    plt.figure(figsize=[30,20])
    plt.bar(data[x_axis], data[y_axis])
    plt.xticks([x for x in range(len(data[x_axis]))], [x for x in data[x_axis].values.tolist()], rotation=90) 
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()

def bins(row, col, min_, max_, width, bins_=[]):
  bins = []
  if len(bins_)==0:
    bins = [[i, i+width] for i in range(min_, max_, width)]
  else:
    for i in range(len(bins_)-1):
      bins.append([bins_[i], bins_[i+1]])
  bin_brac = ""
  for bin in bins:
    if row[col]>=bin[0] and row[col]<=bin[1]:
      bin_brac = f"{bin[0]}-{bin[1]}"
  return bin_brac
````

4. Price Percentiles

````python
def perc_calc(row, column_name,sec_group_cut_offs):
  sg = row[column_name]
  price = row["Price"]
  grp= None

  if price<=sec_group_cut_offs[sg][0]:
    grp = "Low"
  elif (price >= sec_group_cut_offs[sg][0]) and (price < sec_group_cut_offs[sg][1]) :
    grp = "Medium"
  else:
    grp = "High"
  return grp
````

````python
def price_percentile(column_name,new_column_name):
  global df
  df_price_percentile= df.groupby([column_name,'event_date']).mean('Price').reset_index()
  col_cut_offs={}
  for col in df_price_percentile[column_name].unique():
    col_cut_offs[col]=[np.percentile(df_price_percentile[df_price_percentile[column_name]== col]['Price'], 33),
                                np.percentile(df_price_percentile[df_price_percentile[column_name]== col]['Price'], 66)]

  df_price_percentile[new_column_name] = np.nan
  df_price_percentile[new_column_name] = df_price_percentile.apply(lambda x: perc_calc(x,column_name,col_cut_offs),1)
  
  df=pd.merge(df,df_price_percentile[[column_name,'event_date',new_column_name]], on =[column_name,'event_date'], how='left')    
````

````python
df['section_group']=(df['section_name']/100).apply(np.floor)
price_percentile('section_group','section_group_segment')
price_percentile('section_name','section_name_segment')
price_percentile('row_name','row_segment')
price_percentile('SeatNum','seat_segment')
````

````python
def pivot(train_data, group_cols, agg_col):
    data = train_data.groupby(group_cols).agg({agg_col: "count"}).reset_index()
    data = pd.pivot_table(data, values=agg_col, index=[group_cols[0], group_cols[1]], columns=[group_cols[2]], aggfunc=np.sum).reset_index().rename_axis(None, axis=1)
    data = data.fillna(0)
    data["count"] = data["Yes"] + data["No"]
    data["%Attended"] = 100*data["Yes"]/(data["Yes"]+data["No"])
    return data
````

````python
event_names_grp = pivot(df, ["event_name", "event_date", "isAttended"], "acct_id")
event_names_grp = event_names_grp.sort_values("event_date").reset_index(drop=True)
event_names_grp
````
